# Analyse stat

```python

import matplotlib.pyplot as plt
import numpy as np
```

```python
i=1
x,y = np.load('/data/thomas/classification epilepsy/chb-mit-scalp-eeg-database-1.0.0/dataset/temporel'+str(i)+'_x.npy'),np.load('/data/thomas/classification epilepsy/chb-mit-scalp-eeg-database-1.0.0/dataset/temporel'+str(i)+'_y.npy')[:,0]
```

```python
index_1 = np.argwhere(y==1)

index_0 = np.argwhere(y==0)

x_stat_1 = np.sum(np.float32(np.abs(x)),axis = 1)[index_1]/256
x_stat_0 = np.sum(np.float32(np.abs(x)),axis = 1)[index_0]/256
```

```python
plt.figure(dpi=300)
plt.subplot(4,4,1)
for i in range(16):
    plt.subplot(4,4,i+1)
    plt.violinplot(x_stat_1[:,0,i],showmeans=True,showextrema=False)
    plt.violinplot(x_stat_0[:,0,i],showmeans=True,showextrema=False)
    plt.yscale('log')
plt.show()
```

```python
index_1 = np.argwhere(y==0)


x_stat = np.sum(np.float32(np.abs(x)),axis = 1)[index_1]/256
        
for i in range(16):
    plt.violinplot(x_stat[:,0,i],showmeans=True)
    plt.legend([str(i)])
    plt.yscale('log')
    plt.show()
```
